```python
import numpy as np
# 1
# np.full - משמש ליצירת מערך ממולא בערך ספציפי
array_full = np.full((3, 3), 7)
print("Array created with np.full:")
print(array_full)
```

```python
# 2
# a
array_ones = np.ones((4, 4))
print("Array of ones (4x4):")
print(array_ones)

# b
array_threes = np.full((3, 12), 3)
print("Array of threes (3x12):")
print(array_threes)

# c
rows, cols = np.random.randint(2, 6, size=2)
array_random = np.random.randint(11, 23, (rows, cols))
print(f"Random array ({rows}x{cols}):")
print(array_random)
```

```python
# 3
import matplotlib.pyplot as plt
from skimage import io
from PIL import Image

# קריאת התמונה
image = io.imread('jinny.jpg')

# a. הדפסת גודל התמונה
print("Image shape:", image.shape)

# b. סיבוב התמונה עם transpose (סידור צירים באופן ידני)
transposed_image = image.transpose(1, 0, 2)
plt.imshow(transposed_image)
plt.show()

# c. סיבוב התמונה ב-120 מעלות עם PIL
pil_image = Image.open('jinny.jpg')
rotated_image = pil_image.rotate(120)
plt.imshow(rotated_image)
plt.show()

# d. הדפסת 50 העמודות הראשונות
print("First 50 columns of the image:")
print(image[:, :50])
```

```python
# 4
np.random.seed(80)
random_array = np.random.randint(1, 101, (4, 4))
print("Random array (4x4):")
print(random_array)

# a
print("Sum of columns:", np.sum(random_array, axis=0))
print("Sum of rows:", np.sum(random_array, axis=1))

# b
print("Max of columns:", np.max(random_array, axis=0))
print("Max of rows:", np.max(random_array, axis=1))

# c
print("Min of columns:", np.min(random_array, axis=0))
print("Min of rows:", np.min(random_array, axis=1))

# d
print("Mean of columns:", np.mean(random_array, axis=0))
print("Mean of rows:", np.mean(random_array, axis=1))

# e
print("Numbers greater than 50:")
print(random_array[random_array > 50])
```
